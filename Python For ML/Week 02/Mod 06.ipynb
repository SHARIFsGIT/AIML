```python
lst = [x for x in range(500)]

def data_loader(chunk_size, lst):
  for i in range(0, len(lst), chunk_size):
    yield lst[i: i+chunk_size]

x = data_loader(5, lst)

print(next(x))
print(next(x))
```

[G. Max and MIN](https://codeforces.com/group/MWSDmqGsZm/contest/223205/problem/G)

```python
from functools import reduce

n = int(input())
array = list(map(int, input().split()))

mini = reduce(lambda x, y: x if x < y else y, array)
maxi = reduce(lambda x, y: x if x > y else y, array)

print(mini, maxi)
```

[J. Average](https://codeforces.com/group/MWSDmqGsZm/contest/223205/problem/J)

```python
from functools import reduce

n = int(input())

arr = list(map(float, input().split()))

avg = reduce(lambda x,y: x+y, arr)/n

print(avg)
```

```python
n = int(input())

A = list(map(int, input().split()))
B = list(map(int, input().split()))

concat = lambda A, B: B+A

print(*concat(A, B))
```

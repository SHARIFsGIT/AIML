```python
# You must run this code and use that data set for all the questions
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

> Q1. Create NumPy arrays from the ages and salaries data generated above. Print both arrays and display their dtype, ndim, shape, and size



```python
ages = DF['age'].to_numpy()
salaries = DF['salary'].to_numpy()

print(f"Ages Array:\n {ages}\n")
print(f"Salaries Array:\n {salaries}\n")

print(f"Ages Array Dtype: {ages.dtype}\n")
print(f"Salaries Array Dtype: {salaries.dtype}\n")

print(f"Ages Array ndim: {ages.ndim}\n")
print(f"Salaries Array ndim: {salaries.ndim}\n")

print(f"Ages Array shape: {ages.shape}\n")
print(f"Salaries Array shape: {salaries.shape}\n")

print(f"Ages Array size: {ages.size}\n")
print(f"Salaries Array size: {salaries.size}\n")
```

> Q2. Using the salaries array, find and print the highest and lowest salary values. Also, calculate and print the average salary and age using NumPy functions.

```python
highest_salary = np.max(salaries)
lowest_salary = np.min(salaries)

average_salary = np.mean(salaries)
average_age = np.mean(ages)

print(f"Highest Salary: {highest_salary}")
print(f"Lowest Salary: {lowest_salary}")

print(f"Average Salary: {average_salary}")
print(f"Average Age: {average_age}")
```

> Q3. From the ages array, filter and print all ages greater than 30 using a boolean condition. Then count and print how many employees are older than 30.

```python
ages_greater_than_30 = ages[ages > 30]
print(f"Ages greater than 30: {ages_greater_than_30}")

count_older_than_30 = np.sum(ages > 30)
print(f"Number of employees older than 30: {count_older_than_30}")
```



> Q4. Create a new NumPy array that increases every employeeâ€™s age by 5 years (without modifying the original array). Print the new updated ages array.



```python
new_ages = ages + 5

print(f"Original Ages Array: {ages}")
print(f"New Ages Array: {new_ages}")
```



> Q5. Using NumPy, calculate the total salary expense (sum of all salaries) and the difference between the maximum and minimum salary. Print both results.



```python
total_salary_expense = np.sum(salaries)
salary_range = np.max(salaries) - np.min(salaries)

print(f"Total Salary Expense: {total_salary_expense}")
print(f"Salary Difference (Max & Min): {salary_range}")
```



> Q6. Load the employees.csv file into a Pandas DataFrame. Display the first 5 rows, and check basic info using info() and summary statistics using describe().



```python
employees = pd.read_csv('employees.csv')

print("First 5 Rows:")
print(employees.head())

print("\nDataFrame Info:")
print(employees.info())

print("\nSummary Statistics:")
print(employees.describe())
```



> Q7. Display only the id, age, and salary columns. Then show the last 3 rows using tail().



```python
columns = ['id', 'age', 'salary']
selected_columns = employees[columns]

print("Selected Columns:")
print(selected_columns)

print("\nLast 3 Rows:")
print(employees.tail(3))
```



> Q8. Filter the DataFrame to show only employees who work in the IT department. Print the result and show the total number of IT employees.



```python
it_employees = employees.loc[employees['dept'] == 'IT']

print("IT Employees:")
print(it_employees)

print("\nTotal Number of IT Employees:", it_employees.shape[0])
```



> Q9. Sort the DataFrame by salary in descending order and display the top 3 highest-paid employees along with their department and age.



```python
sorted_employees = employees.sort_values(by='salary', ascending=False)

top_3_employees = sorted_employees.head(3)[['salary', 'dept', 'age']]

print("Top 3 Highest-Paid Employees:")
print(top_3_employees)

```



> Q10. Replace all salary values greater than 80000 with 80000 using loc. Then calculate and print the new average salary of all employees after replacement.



```python
employees.loc[employees['salary'] > 80000, 'salary'] = 80000

new_average_salary = employees['salary'].mean()

print("New Average Salary:", new_average_salary)
```

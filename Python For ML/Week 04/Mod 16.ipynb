Dataset : [Python-for-ML/student_IQdata.csv at main Â· phitronio/Python-for-ML](https://github.com/phitronio/Python-for-ML/blob/main/student_IQdata.csv)


1. Load the dataset from the URL into a pandas DataFrame.
2. Create a scatter plot with Study_Hours on the X-axis and IQ on the Y-axis with seaborn
3. Create a scatter plot with Shoe_Size on the X-Axis and Study_Hours on the Y-axis with seabron
4. Create a histogram of Shoe_Size using plotly
5. After you plot it, answer the question: Does there appear to be a correlation?


```python
df = pd.read_csv('https://raw.githubusercontent.com/phitronio/Python-for-ML/main/student_IQdata.csv')
df
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Study_Hour', y='IQ_Score', s=100, alpha=0.6, color='steelblue')
plt.title('Study Hours vs IQ Score', fontsize=16, fontweight='bold')
plt.xlabel('Study Hours', fontsize=12)
plt.ylabel('IQ Score', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Shoe_Size', y='Study_Hour', s=100, alpha=0.6, color='coral')
plt.title('Shoe Size vs Study Hours', fontsize=16, fontweight='bold')
plt.xlabel('Shoe Size', fontsize=12)
plt.ylabel('Study Hours', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
fig = px.histogram(df, x='Shoe_Size', nbins=15, title='Distribution of Shoe Sizes', color_discrete_sequence=['indianred'])

fig.update_layout(title_font_size=16, xaxis_title='Shoe Size', yaxis_title='Frequency', showlegend=False, height=500)

fig.show()
```

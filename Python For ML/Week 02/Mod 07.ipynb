```python
from functools import reduce

with open("./sample_data/sample.txt", "r") as file:
  lines = file.readlines()
  print(lines)

  # Number of lines
  total_lines = len(lines)
  print(total_lines)

  # Number of words
  words = list(map(lambda x: len(x.split()), lines))
  print(words)
  total_words = reduce(lambda x, y: x + y, words)
  print(total_words)

  # Number of characters
  print(lines)
  lines = list(map(str.strip, lines))
  print(lines)

  lines = list(map(lambda x: x.replace(" ", ""), lines))
  print(lines)

  chars = list(map(lambda x: len(x), lines))
  print(chars)

  total_chars = reduce(lambda x, y: x + y, chars)
  print(total_chars)

with open("./sample_data/result.txt", "w") as file:
  file.write(f"Total lines: {total_lines}\n")
  file.write(f"Total words: {total_words}\n")
  file.write(f"Total characters: {total_chars}")
```

Create a text file named emails.txt containing multiple email addresses, some in uppercase and some repeated.
Write a Python program to clean the data by converting all emails to lowercase, removing duplicates, sorting them alphabetically, and saving the cleaned list to a new file named unique_emails.txt.

```python
with open("./sample_data/emails.txt", "r") as file:
  emails = file.readlines()
  emails = list(map(str.strip, emails))
  emails = list(map(str.lower, emails))

  unique_emails = list(set(emails))
  unique_emails.sort()

  print(unique_emails)

with open("./sample_data/unique_emails.txt", "w") as file:
  for email in unique_emails:
    file.write(email + "\n")
```

Create a text file named names.txt containing a list of names.
Some names may have extra spaces, inconsistent capitalization (like “aLiCe” or “ bob ”), or appear more than once.
Write a Python program to clean the data by:
1.  Removing extra spaces.
2.  Converting all names so that only the first letter is capitalized.
3.  Removing duplicate names.
4. Sorting the names alphabetically.
Finally, save the cleaned list into a new file named clean_names.txt.

```python
with open("./sample_data/names.txt", "r") as file:
  names = file.readlines()

names = [name.strip().capitalize() for name in names if name.strip() != ""]
unique_names = sorted(set(names))

with open("./sample_data/clean_names.txt", "w") as file:
    for name in unique_names:
        file.write(name + "\n")
```

Write a Python program that creates a list of 5 numbers.
Ask the user for an index number and print the value at that index.
If the index is out of range, handle the exception and print "Invalid index!".

```python
nums = [10, 20, 30, 40, 50]

try:
  index = int(input("Enter an index: "))
  print(f"Value at index: {index} is {nums[index]}")
except IndexError:
  print(f"There is no value at index: {index}")
```

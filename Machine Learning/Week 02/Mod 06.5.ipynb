## Section 7 — Target Distribution and Class Imbalance

Understand how balanced or imbalanced your target classes are.


```python
# 1. Create a countplot of the target column
# A countplot shows the number of occurrences for each class in the target variable
# This is essential for understanding class balance in classification problems 

print("Target Variable Distribution - Countplot")
plt.figure(figsize=(8, 6))

# Create countplot
sns.countplot(x=target_column, data=df, hue=df[target_column], palette=['salmon', 'lightgreen'])
plt.title('Distribution of Survival (Target Variable)', fontsize=14, fontweight='bold')
plt.xlabel('Survived (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Add value labels on top of bars
for container in plt.gca().containers:
    plt.gca().bar_label(container)

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# 2. Show value counts and normalized proportions for the target
# Value counts show absolute numbers
# Normalized proportions show percentages - critical for identifying imbalance

print("\n" + "="*50)
print("Target Variable - Absolute Counts:")
print("="*50)
survival_counts = df[target_column].value_counts()
print(survival_counts)
print()

print("="*50)
print("Target Variable - Proportions (Percentages):")
print("="*50)
survival_proportions = df[target_column].value_counts(normalize=True) * 100
print(survival_proportions.round(2))
print()

# Create a more detailed summary
print("="*50)
print("Summary:")
print("="*50)
print(f"Total Passengers: {len(df)}")
print(f"Survived (1): {survival_counts[1]} passengers ({survival_proportions[1]:.2f}%)")
print(f"Did Not Survive (0): {survival_counts[0]} passengers ({survival_proportions[0]:.2f}%)")
print(f"Survival Rate: {survival_proportions[1]:.2f}%")
print(f"Mortality Rate: {survival_proportions[0]:.2f}%")
```

### Student Notes and Observations

**Class Distribution:**
- Did Not Survive (0): 549 passengers (approximately 61.6%)
- Survived (1): 342 passengers (approximately 38.4%)

**Class Imbalance Analysis:**
The dataset shows a moderate class imbalance with about a 60:40 ratio favoring non-survivors. This reflects the tragic historical reality of the Titanic disaster where the majority of passengers did not survive.

**Implications for Machine Learning:**

1. **Degree of Imbalance**: 
   - The imbalance is noticeable but not extreme (would be extreme if it were 90:10 or worse)
   - This level of imbalance (60:40) is manageable but should be considered

2. **Potential Model Bias**:
   - Without proper handling, models might be biased toward predicting "Did Not Survive" (class 0)
   - A naive model that always predicts "0" would be correct 61.6% of the time
   - We need to ensure the model truly learns patterns, not just predicts the majority class

3. **Evaluation Metrics Consideration**:
   - Accuracy alone won't be sufficient to evaluate model performance
   - Should use additional metrics like Precision, Recall, F1-Score, and ROC-AUC
   - Confusion matrix will be important to see performance on both classes

4. **Preprocessing Strategies for Module 07**:
   - Could use techniques like SMOTE (Synthetic Minority Over-sampling)
   - Could apply class weights during model training
   - Could use stratified sampling when splitting train/test data
   - For now, the imbalance is moderate enough that we might proceed without resampling

**Historical Context**: 
The survival rate of 38.4% aligns with historical records of the Titanic disaster, adding credibility to the dataset.

## Section 11 — Categorical Features versus Target (Crosstab and Stacked Bar Plot)

Examine how target proportions change across categories.


```python
# For each categorical feature:
# 1. Create a crosstab between the feature and target with normalize="index"
# 2. Plot a stacked bar chart of the proportions
# 3. Comment on how the target distribution changes across categories

# Crosstab shows how survival rates differ across different categories
# normalize="index" shows proportions within each category (rows sum to 1)
# This helps us see which categories had higher/lower survival rates

print("Analyzing Categorical Features vs Target (Survival)\n")

# Iterate through each categorical feature
for col in categorical_features:
    print("="*70)
    print(f"Analysis: {col} vs Survived")
    print("="*70)
    
    # 1. Create crosstab with normalized values
    # normalize="index" gives us the proportion of survivors/non-survivors within each category
    crosstab = pd.crosstab(df[col], df[target_column], normalize='index')
    
    # Also show absolute counts for reference
    crosstab_counts = pd.crosstab(df[col], df[target_column])
    
    print(f"\nAbsolute Counts:")
    print(crosstab_counts)
    print(f"\nProportions (Survival Rate by {col}):")
    print(crosstab)
    print()
    
    # Calculate survival rates for each category
    survival_rates = crosstab[1] * 100  # Convert to percentage
    print(f"Survival Rate by {col}:")
    for category in survival_rates.index:
        print(f"  {category}: {survival_rates[category]:.2f}%")
    print()
    
    # 2. Plot stacked bar chart
    fig, ax = plt.subplots(1, 2, figsize=(14, 5))
    
    # Stacked bar chart showing proportions
    crosstab.plot(kind='bar', stacked=True, ax=ax[0], 
                  color=['salmon', 'lightgreen'], 
                  edgecolor='black', alpha=0.8)
    ax[0].set_title(f'Survival Proportions by {col}', fontsize=12, fontweight='bold')
    ax[0].set_xlabel(col, fontsize=11)
    ax[0].set_ylabel('Proportion', fontsize=11)
    ax[0].set_ylim([0, 1])
    ax[0].legend(['Did Not Survive (0)', 'Survived (1)'], loc='upper left')
    ax[0].tick_params(axis='x', rotation=45)
    ax[0].grid(axis='y', alpha=0.3)
    
    # Add percentage labels on the bars
    for container in ax[0].containers:
        ax[0].bar_label(container, fmt='%.2f', label_type='center')
    
    # Grouped bar chart showing survival rates for easier comparison
    survival_rates.plot(kind='bar', ax=ax[1], color='lightgreen', 
                       edgecolor='black', alpha=0.8)
    ax[1].set_title(f'Survival Rate by {col}', fontsize=12, fontweight='bold')
    ax[1].set_xlabel(col, fontsize=11)
    ax[1].set_ylabel('Survival Rate (%)', fontsize=11)
    ax[1].tick_params(axis='x', rotation=45)
    ax[1].grid(axis='y', alpha=0.3)
    
    # Add value labels on bars
    for container in ax[1].containers:
        ax[1].bar_label(container, fmt='%.1f%%')
    
    plt.tight_layout()
    plt.show()
    
    print("\n")
```

```python
nested_squares = [[x**2 for x in range(i, i+5) if x%2 == 0] for i in range(1, 10, 3)]
print("Squares: ", nested_squares)
```

```python
number_description = {
    num: ("Even and divisible by 3" if num%2 == 0 and num%3 == 0
        else "Even" if num%2 == 0
        else "Odd and divisible by 3" if num%3 == 0
        else "Odd"
      ) for num in range (1, 11)
}
print("Number description: ", number_description)
```

```python
users_data = [
    {"user_id": 1, "name": "Alice", "orders": [{"order_id": 101, "amount": 250}, {"order_id": 102, "amount": 450}]},
    
    {"user_id": 2, "name": "Bob", "orders": [{"order_id": 103, "amount": 150}]},
    
    {"user_id": 3, "name": "Charlie", "orders": []},
    
    {"user_id": 4, "name": "David", "orders": [{"order_id": 104, "amount": 300}, {"order_id": 105, "amount": 200}, {"order_id": 106, "amount": 400}]}
]

print("Total amounts spent by each user: ", {user["name"]: sum(order["amount"] for order in user["orders"]) for user in users_data})
```

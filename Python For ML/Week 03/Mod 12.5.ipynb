Create a Pandas Series from the list [10, 20, 30, 40, 50]. Print the Series, its index, and data type. Then create a DataFrame with columns "Name", "Age", and "City" containing at least 4 rows of data

```python
import pandas as pd

# Create a Pandas Series from a list
data = [10, 20, 30, 40, 50]
series = pd.Series(data)

print("Pandas Series:")
print(series)
print("-" * 40)

# Print index
print("Index:", series.index)
# Print data type
print("Data type:", series.dtype)
print("-" * 40)

# Create a DataFrame with columns Name, Age, and City
data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age': [25, 30, 28, 35],
    'City': ['Berlin', 'Paris', 'Rome', 'Madrid']
}

df = pd.DataFrame(data_dict)

print("Pandas DataFrame:")
print(df)
print("-" * 40)
```

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan'],
    'Age': [25, 30, 28, 35, 22],
    'Salary': [50000, 60000, 55000, 70000, 48000],
    'City': ['Berlin', 'Paris', 'Rome', 'Madrid', 'Lisbon']
}

df = pd.DataFrame(data)

# Print number of rows and columns
print("Number of Rows and Columns:")
print(df.shape)
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
print("-" * 50)

# Print column data types
print("Column Data Types:")
print(df.dtypes)
print("-" * 50)

# Print summary statistics
print("Summary Statistics:")
print(df.describe())
print("-" * 50)
```

Print .columns, .index. Access a single column and one full row.

```python
import pandas as pd

# Create the DataFrame
data = {
    'Product': ['Pen', 'Book', 'Eraser', 'Pencil'],
    'Price': [10, 50, 5, 7],
    'Quantity': [100, 60, 200, 150]
}

df = pd.DataFrame(data)

# Print columns and index
print("Columns:", df.columns)
print("Index:", df.index)
print("-" * 40)

# Access a single column (e.g., 'Price')
print("Single Column - 'Price':")
print(df['Price'])
print("-" * 40)

# Access one full row (e.g., row at index 1)
print("Full Row at index 1:")
print(df.loc[1])   # or df.iloc[1]
print("-" * 40)
```

Using .loc[] and .iloc[], show the first row, the second column, and a 2×2 block    from the middle of your DataFrame.

```python
import pandas as pd

# Create the DataFrame again
data = {
    'Product': ['Pen', 'Book', 'Eraser', 'Pencil'],
    'Price': [10, 50, 5, 7],
    'Quantity': [100, 60, 200, 150]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
print("-" * 50)

# First row using .loc[] and .iloc[]
print("First Row (.loc):")
print(df.loc[0])      # by label
print("-" * 30)

print("First Row (.iloc):")
print(df.iloc[0])     # by integer position
print("-" * 50)

# Second column using .loc[] and .iloc[]
print("Second Column (.loc):")
print(df.loc[:, 'Price'])   # select column by name
print("-" * 30)

print("Second Column (.iloc):")
print(df.iloc[:, 1])        # select column by position
print("-" * 50)

# 2×2 block from the middle using .loc[] and .iloc[]
# Middle rows = index 1 & 2, middle columns = 'Price' and 'Quantity'
print("2×2 Block (.loc):")
print(df.loc[1:2, ['Price', 'Quantity']])
print("-" * 30)

print("2×2 Block (.iloc):")
print(df.iloc[1:3, 1:3])   # rows 1–2, columns 1–2
print("-" * 50)
```

Create a 4×3 DataFrame of student names and marks. Drop one column using .drop(), drop one row by index

```python
import pandas as pd

# Create a 4×3 DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Math Marks': [85, 92, 78, 88],
    'Science Marks': [90, 84, 80, 95]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print("-" * 50)

# Drop one column (e.g., 'Science Marks')
df_dropped_col = df.drop(columns=['Science Marks'])
print("After Dropping 'Science Marks' Column:")
print(df_dropped_col)
print("-" * 50)

# Drop one row (e.g., row with index 2)
df_dropped_row = df.drop(index=2)
print("After Dropping Row with Index 2:")
print(df_dropped_row)
print("-" * 50)
```

Create a DataFrame of employees with columns "Name", "Department", and "Salary". Sort it by "Salary" in ascending order and by "Name" in descending order.

```python
import pandas as pd

# Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan'],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR'],
    'Salary': [50000, 70000, 65000, 70000, 48000]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print("-" * 50)

# Sort by Salary (ascending) and then by Name (descending)
sorted_df = df.sort_values(by=['Salary', 'Name'], ascending=[True, False])

print("Sorted DataFrame (by Salary ↑ and Name ↓):")
print(sorted_df)
print("-" * 50)
```

Create a DataFrame of 8 students with columns "Name" and "Score". Filter and print students with Score > 60 and those with Score < 40

```python
import pandas as pd

# Create a DataFrame of 8 students
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan', 'Fiona', 'George', 'Hannah'],
    'Score': [75, 35, 82, 58, 92, 40, 28, 65]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print("-" * 50)

# Filter students with Score > 60
high_scores = df.loc[df['Score'] > 60]
print("Students with Score > 60:")
print(high_scores)
print("-" * 50)

# Filter students with Score < 40
low_scores = df.loc[df['Score'] < 40]
print("Students with Score < 40:")
print(low_scores)
print("-" * 50)
```

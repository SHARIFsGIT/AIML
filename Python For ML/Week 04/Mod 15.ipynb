```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hours = [2, 3, 2, 5, 1, 3, 6]
days = [1, 2, 3, 4, 5, 6, 7]

plt.plot(days, hours)

plt.xlabel('Days')
plt.ylabel('Hours')
plt.title('Study per day')
plt.show()
```

```python
hours1 = [2, 3, 2, 5, 1, 3, 6]
hours2 = [6, 4, 6, 2, 4, 3, 2]
days = [1, 2, 3, 4, 5, 6, 7]

plt.plot(days, hours1, label='Hours Set 1', color='blue', marker='o')
plt.plot(days, hours2, label='Hours Set 2', color='red', marker='x')

plt.xlabel('Days')
plt.ylabel('Hours')
plt.title('Daily Study Hours Comparison')
plt.legend()
plt.grid(True)
plt.show()
```

```python
students = pd.read_csv('./sample_data/student_IQdata.csv')

students
```

```python
plt.figure(figsize=(10, 7))
scatter = plt.scatter(students['Study_Hour'], students['IQ_Score'],
                      c=students['Chilling_Hours'], # Color based on 'Chilling_Hours'
                      s=students['Shoe_Size'] * 10, # Size based on 'Shoe_Size' (scaled for visibility)
                      cmap='plasma', # Colormap for the 'Chilling_Hours'
                      alpha=0.7, # Transparency for better visibility of overlapping points
                      edgecolors='w', # White edges for markers
                      linewidth=0.5)

plt.xlabel('Study Hours')
plt.ylabel('IQ Score')
plt.title('Study Hours vs. IQ Score (Colored by Chilling Hours, Sized by Shoe Size)')

cbar = plt.colorbar(scatter)
cbar.set_label('Chilling Hours')

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 7))

# 'bins' determines the number of bars
# 'density=True' normalizes the bars so their total area sums to 1 (if you want to compare distributions)
hist_values, bin_edges, _ = plt.hist(students['IQ_Score'], bins=20, density=True,
                                     color='#74A089', edgecolor='w', alpha=0.8,
                                     label='IQ Score Histogram')

plt.xlabel('IQ Score', fontsize=14)
plt.ylabel('Frequency / Density', fontsize=14)
plt.title('Distribution of IQ Scores', fontsize=16, weight='bold', color='#333333')

plt.grid(axis='y', linestyle=':', alpha=0.7, color='gray')

plt.legend(fontsize=10, frameon=True, shadow=True, borderpad=1) # Add a fancier legend for the histogram

plt.tight_layout() # Adjust plot parameters for a tight layout
plt.show()
```

```python
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female'],
    'Study_Hour': [4, 6, 3, 5, 2],
    'Attendance': [95, 78, 88, 100, 65],
    'Grade': ['A', 'B', 'C', 'A', 'D']
}

df = pd.DataFrame(data)

df
```

```python
gender_group = df.groupby('Gender').size()

plt.figure(figsize=(8, 6))

plt.bar(gender_group.index, gender_group.values,
        color=['#1f77b4', '#ff7f0e'],
        edgecolor='black',
        alpha=0.8)

plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.title('Distribution of Students by Gender', fontsize=14, weight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```

```python
df = pd.read_csv('./sample_data/student_data.csv')

df
```

```python
location_counts = df.groupby('Location').size().sort_values(ascending=False)

plt.figure(figsize=(10, 6))

plt.bar(location_counts.index, location_counts.values, color='#8ECFC9', edgecolor='black', alpha=0.8)

plt.xlabel('Location', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.title('Distribution of Students by Location', fontsize=14, weight='bold')

plt.xticks(rotation= 45, ha='right') # Rotate x-axis labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
completion_status_counts = df.groupby('CompletionStatus').size().sort_values(ascending=False)

plt.figure(figsize=(9, 6))

colors_status = {'Completed': 'g', 'In Progress': 'y', 'Not Started': 'r'}

bar_colors_status = [colors_status.get(status) for status in completion_status_counts.index]

plt.bar(completion_status_counts.index, completion_status_counts.values,
        color=bar_colors_status, # Apply custom colors
        edgecolor='black',
        alpha=0.8)

plt.xlabel('Completion Status', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.title('Distribution of Students by Completion Status', fontsize=14, weight='bold')

plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
completion_status_counts = df.groupby('CompletionStatus').size().sort_values(ascending=False)

plt.figure(figsize=(9, 9))

colors_status = {'Completed': 'g', 'In Progress': 'y', 'Not Started': 'r'}
pie_colors = [colors_status.get(status) for status in completion_status_counts.index]

explode_values = [0.1 if i == 0 else 0 for i in range(len(completion_status_counts))]

plt.pie(completion_status_counts.values,
        labels=completion_status_counts.index,
        colors=pie_colors,
        autopct='%1.1f%%',
        startangle=90,
        wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'antialiased': True},
        explode=explode_values,
        shadow=True,
        textprops={'fontsize': 13, 'weight': 'bold', 'color': 'black'})

plt.title('Distribution of Students by Completion Status', fontsize=16, weight='bold', color='#333333')

plt.axis('equal')
plt.tight_layout()
plt.show()
```

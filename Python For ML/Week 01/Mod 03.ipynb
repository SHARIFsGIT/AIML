```python
n = int(input())
arr_string = input().split()

arr = []

for i in range(n):
    a = int(arr_string[i])
    arr.append(a)

min_value = arr[0]
position = 1

for i in range(1, n):
    if arr[i] < min_value:
      min_value = arr[i]
      position = i+1

print(min_value, position)
```

```python
n = int(input())
arr_string = input().split()

arr = []

for i in range(n):
  a = int(arr_string[i])
  arr.append(a)

min_index = arr.index(min(arr))
max_index = arr.index(max(arr))

arr[min_index], arr[max_index] = arr[max_index], arr[min_index]

for num in arr:
  print(num, end=" ")
```

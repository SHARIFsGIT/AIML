### Student Marks Analyzer

You are helping a teacher check exam results.
 Create a NumPy array with 10 random integers between 0 and 100 (student scores).

 Find:
 - the highest score
 - the lowest score
 - the average score

 Hint: use np.random.randint(), arr.max(), arr.min(), arr.mean()

```python
import numpy as np

scores = np.random.randint(0, 101, size = 10)

print("Student Scores: ", scores)

highest_score = scores.max()
lowest_score = scores.min()
average_score = scores.mean()

print("Highest Score: ", highest_score)
print("Lowest Score: ", lowest_score)
print("Average Score: ", average_score)
```

### Warehouse Weights

A warehouse stores item weights (in kg):
 arr = np.array([45, 60, 55, 30, 90, 75, 40, 50])

 Find and print:
 - all items heavier than 50 kg
 - the sorted weights
 - the index of the heaviest item

 Hint: use arr[arr > 50], np.sort(arr), arr.argmax()

```python
arr = np.array([45, 60, 55, 30, 90, 75, 40, 50])

heavier_items = arr[arr > 50]
sorted_weights = np.sort(arr)
heaviest_item_index = arr.argmax()

print("All items heavier than 50 kg: ", heavier_items)
print("Sorted weights: ", sorted_weights)
print("Index of the heaviest item: ", heaviest_item_index)
```

### Temperature Report

A weather station collects temperatures (°C) from 4 zones.
 Generate random integers between 20 and 40 for a 4×4 grid.

 Then find:
 - the average temperature of each zone (row)
 - the highest temperature in the grid
 - replace all temperatures above 35 with 35

```python
temps = np.random.randint(20, 41, size=(4, 4))

print("Temperature Grid (°C):\n", temps)

average_temps = np.mean(temps, axis=1)
highest_temp = np.max(temps)
temps[temps > 35] = 35

print("\nAverage Temperatures (°C):", average_temps)
print("Highest Temperature (°C):", highest_temp)
print("\nTemperature Grid after Replacement (°C):\n", temps)
```

### Traffic Data
A city records the number of cars passing 5 intersections in 3 time slots (morning, afternoon, night).
 Create a 3×5 array of random integers between 0 and 50.

 Then find:
 - total cars per intersection (sum by column)
 - total cars per time slot (sum by row)

```python
traffic = np.random.randint(0, 51, size=(3, 5))

print("Traffic Data:\n", traffic)

total_cars_per_intersection = np.sum(traffic, axis=0)
total_cars_per_time_slot = np.sum(traffic, axis=1)

print("\nTotal Cars per Intersection:", total_cars_per_intersection)
print("Total Cars per Time Slot:", total_cars_per_time_slot)
```

### Fruit Store Stock
You sell fruits in your store. Stock levels (kg) for 5 days are:
 stock = np.array([12, 15, 9, 20, 18])

 A new delivery adds 5 kg to each stock.
 Then 10% of all fruits get spoiled.

 Find the final stock levels.

 Hint: use broadcasting and simple math.


```python
stock = np.array([12, 15, 9, 20, 18], dtype=float)

new_delivery = 5
stock += new_delivery

print("New Delivery (kg):", new_delivery)
print("Updated Stock Levels (kg):", stock)

spoiled_fruits = stock * 0.1
stock -= spoiled_fruits

print("Final Stock Levels (kg):", stock)
```

### Shop Sales Data
You run a small electronics shop. The data is stored like this:

 data = {
 	'Product': ['Phone', 'Laptop', 'Tablet', 'Monitor', 'Mouse'],
 	'Price': [700, 1200, 300, 250, 50],
 	'Units_Sold': [10, 5, 8, 6, 20]
 }
 df = pd.DataFrame(data)

 Tasks:
 1. Display only the 'Product' and 'Price' columns.
 2. Access the 3rd row using both iloc and loc.
 3. Show the top 3 products with the highest revenue.

```python
data = {
    'Product': ['Phone', 'Laptop', 'Tablet', 'Monitor', 'Mouse'],
    'Price': [700, 1200, 300, 250, 50],
    'Units_Sold': [10, 5, 8, 6, 20]
}

df = pd.DataFrame(data)

# Display only 'Product' and 'Price' columns
print("Product and Price columns:")
print(df[['Product', 'Price']])

# Access the 3rd row using both iloc and loc
print("\n3rd row using iloc:")
print(df.iloc[2])   # index-based access

print("\n3rd row using loc:")
print(df.loc[2])    # label-based access (same as iloc here, since index = 0,1,2,...)

# Show the top 3 products with the highest revenue
df['Revenue'] = df['Price'] * df['Units_Sold']
top3 = df.sort_values(by='Revenue', ascending=False).head(3)

print("\nTop 3 products with highest revenue:")
print(top3[['Product', 'Revenue']])
```

### Attendance Tracker
Here’s an attendance sheet:

 data = {
 	'Student': ['Amin', 'Tina', 'Rafi', 'Nadia', 'Sabbir'],
 	'Days_Present': [20, 18, 22, 15, 25],
 	'Days_Absent': [2, 4, 1, 7, 0]
 }
 df = pd.DataFrame(data)

 Tasks:
 1. Make 'Student' the index column.
 2. Drop the 'Days_Absent' column.


```python
import pandas as pd

data = {
    'Student': ['Amin', 'Tina', 'Rafi', 'Nadia', 'Sabbir'],
    'Days_Present': [20, 18, 22, 15, 25],
    'Days_Absent': [2, 4, 1, 7, 0]
}

# Create DataFrame
df = pd.DataFrame(data)

# Show the original DataFrame
print("Original DataFrame:")
print(df)

# Make 'Student' the index column
df.set_index('Student', inplace=True)

# Drop the 'Days_Absent' column
df.drop('Days_Absent', axis=1, inplace=True)

# Display the final DataFrame
print("\nFinal Attendance DataFrame:")
print(df)
```

### Filtering Employee Data
Employee record:

 data = {
 	'Name': ['Rafiq', 'Nadia', 'Tuhin', 'Rima', 'Faruk'],
 	'Department': ['HR', 'IT', 'IT', 'Sales', 'Sales'],
 	'Salary': [40000, 55000, 52000, 45000, 47000],
 	'Experience': [2, 5, 4, 3, 6]
 }
 df = pd.DataFrame(data)

 Tasks:
 1. Show employees with Salary > 45000.
 2. Show employees from the IT department only.
 3. Show employees with Salary > 45000 and Experience > 3.
 4. Sort the employees by Experience (ascending).

```python
import pandas as pd

# Employee data
data = {
    'Name': ['Rafiq', 'Nadia', 'Tuhin', 'Rima', 'Faruk'],
    'Department': ['HR', 'IT', 'IT', 'Sales', 'Sales'],
    'Salary': [40000, 55000, 52000, 45000, 47000],
    'Experience': [2, 5, 4, 3, 6]
}

df = pd.DataFrame(data)

print("Original Employee Data:")
print(df)

# Employees with Salary > 45000
print("\nEmployees with Salary > 45000:")
print(df[df['Salary'] > 45000])

# Employees from the IT department only
print("\nEmployees from IT Department:")
print(df[df['Department'] == 'IT'])

# Employees with Salary > 45000 and Experience > 3
print("\nEmployees with Salary > 45000 and Experience > 3:")
print(df[(df['Salary'] > 45000) & (df['Experience'] > 3)])

# Sort employees by Experience (ascending)
print("\nEmployees sorted by Experience (ascending):")
print(df.sort_values(by='Experience', ascending=True))
```

### City Weather Dashboard
data = {
 	'City': ['Dhaka', 'Sylhet', 'Chittagong', 'Rajshahi', 'Khulna'],
 	'Temp': [32, 28, 30, 35, 33],
 	'Humidity': [60, 75, 70, 55, 65]
 }
 df = pd.DataFrame(data)

 Tasks:
 1. Show only the cities where FeelsLike > 33.
 2. Sort the DataFrame by FeelsLike (descending).
 3. Find which city has the highest humidity.


```python
import pandas as pd

data = {
    'City': ['Dhaka', 'Sylhet', 'Chittagong', 'Rajshahi', 'Khulna'],
    'Temp': [32, 28, 30, 35, 33],
    'Humidity': [60, 75, 70, 55, 65]
}

df = pd.DataFrame(data)

# Add a new column "FeelsLike" (simple formula combining Temp & Humidity)
# Example formula: FeelsLike = Temp + (Humidity / 100) * 5
df['FeelsLike'] = df['Temp'] + (df['Humidity'] / 100) * 5

print("City Weather Data:")
print(df)

# Show only the cities where FeelsLike > 33
print("\nCities where FeelsLike > 33:")
print(df[df['FeelsLike'] > 33])

# Sort the DataFrame by FeelsLike (descending)
print("\nData sorted by FeelsLike (descending):")
print(df.sort_values(by='FeelsLike', ascending=False))

# Find which city has the highest humidity
highest_humidity_city = df.loc[df['Humidity'].idxmax(), 'City']
print("\nCity with the highest humidity:", highest_humidity_city)
```

```python
n = int(input())

numbers = input().split()

positive_count = 0
negative_count = 0
even_count = 0
odd_count = 0

for i in range(n):
    x = int(numbers[i])
    
    if x > 0:
        positive_count += 1
    elif x < 0:
        negative_count += 1
    
    if x % 2 == 0:
        even_count += 1
    else:
        odd_count += 1
        
print("Even:", even_count)
print("Odd:", odd_count)
print("Positive:", positive_count)
print("Negative:", negative_count)
```

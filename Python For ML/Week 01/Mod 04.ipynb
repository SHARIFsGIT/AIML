[Sensor Calibrator â€” Min-Max Scaling](https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/b3-sensor-calibrator-min-max-scaling)

```python
inp = input().split()

x = float(inp[0])
min_v = float(inp[1])
max_v = float(inp[2])

norm = (x - min_v) / (max_v - min_v)

print(f"{norm:.2f}")
```

[Mini Loss Tracker](https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/b4-mini-loss-tracker-average-monitor)

```python
N = int(input())
target = float(input())

total = 0.0

for i in range(N):
  loss = float(input())
  total += loss

average = total / N

if average <= target:
  print("PASS")
else:
  print("RETRY")
```

[Majority Vote Gate](https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/b5-majority-vote-gate-simple-ensemble)

```python
N = int(input())

yes = 0
no = 0

for i in range(N):
  votes = input().split()

  for vote in votes:
    if vote == "YES":
      yes += 1
    else:
      no += 1

if yes >= no:
  print("ACCEPT")
else:
  print("REJECT")
```

[AI Mood Detector](https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/ai-mood-detector)

```python
message  = input()

if message.find("happy") != -1 or message.find("joy") != -1 or message.find("smile") != -1:
  print("Happy Mood")
elif message.find("sad") != -1 or message.find("cry") != -1 or message.find("angry") != -1:
    print("Sad Mood")
else:
    print("Neutral Mood")
```

[AI Brightness Analyzer](https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/p2-a2-ai-brightness-analyzer)

```python
inp = input().split()

total = 0

for i in range(len(inp)):
  total += int(inp[i])

avg = total / len(inp)

if avg < 85:
  print("Dark Image")
elif avg >= 85 and avg <= 170:
  print("Normal Image")
else:
  print("Bright Image")
```

[AI Bias Detector](https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/p2-a3-ai-bias-detector)

```python
inp = input().split()

a = inp.count("A")
b = inp.count("B")

total = len(inp)

if (a/total > 0.7 )or (b/total > 0.7 ):
  print("Biased Model")
else:
  print("Fair Model")
```

[AI Keyword Counter](https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/p2-a4-ai-keyword-counter)

```python
text = input().split()

keywords = ["ai", "data", "model", "learn", "train", "neural"]
count = 0

for word in text:
    if word in keywords:
        count += 1

if count >= 2:
    print("AI Detected")
else:
    print("Not AI Related")
```

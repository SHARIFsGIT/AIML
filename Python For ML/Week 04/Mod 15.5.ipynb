```python
import pandas as pd

df = pd.read_csv('sample_data/PracticeDay.csv', on_bad_lines='skip')
df
```

```python
df['Discounted_Price'] = df['Price'] * 0.90
df
```

```python
df.to_csv('discounted_products.csv', index=False)
```

```python
print(f"Number of duplicate EmployeeID entries: {df.duplicated(subset='EmployeeID').sum()}")
print(f"DataFrame shape before removing duplicates: {df.shape}")

df.drop_duplicates(subset='EmployeeID', inplace=True)
print(f"DataFrame shape after removing duplicates: {df.shape}")
df
```

```python
print("Missing values before handling:")
print(df[['Age', 'Salary']].isnull().sum())

# Calculate mean of 'Age' column
mean_age = df['Age'].mean()
print(f"\nMean Age: {mean_age:.2f}")

# Fill missing 'Age' values with the mean
df['Age'] = df['Age'].fillna(mean_age)

# Drop rows where 'Salary' is missing
df.dropna(subset=['Salary'], inplace=True)

print("\nMissing values after handling:")
print(df[['Age', 'Salary']].isnull().sum())

print("\nDataFrame after handling missing values:")
df.head()
```

```python
print("Statistical Analysis for Marks:")

# Calculate mean and max for each subject
marks_columns = ['Marks1', 'Marks2', 'Marks3']
for col in marks_columns:
    print(f"\n{col} - Mean: {df[col].mean():.2f}, Max: {df[col].max()}")

# Compute total marks per student
df['Total_Marks'] = df[marks_columns].sum(axis=1)
print("\nDataFrame with 'Total_Marks' column:")
print(df)

# Find standard deviation of total marks
std_dev_total_marks = df['Total_Marks'].std()
print(f"\nStandard Deviation of Total Marks: {std_dev_total_marks:.2f}")
```

```python
import matplotlib.pyplot as plt

# Histogram of Age
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=5, edgecolor='w')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Scatter plot (Age vs Salary)
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Salary'], alpha=0.7)
plt.title('Age vs Salary Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Scatter plot (Price vs Discounted Price)
plt.figure(figsize=(8, 6))
plt.scatter(df['Price'], df['Discounted_Price'], alpha=0.7, color='red')
plt.title('Price vs Discounted Price Scatter Plot')
plt.xlabel('Original Price')
plt.ylabel('Discounted Price')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})

print("\033[95m" + "=== Original DataFrame ===" + "\033[0m")
print(df, "\n")

# ============================================================
# Fill missing values with custom values per column
# ============================================================
df_filled = df.fillna({'A': 0, 'B': 99, 'C': 'Unknown'})

print("\033[94m" + "=== Fill Missing Values with Custom Rules ===" + "\033[0m")
print(df_filled, "\n")

# ============================================================
# Forward Fill only Column 'B'
# ============================================================
df_ffill = df.copy()
df_ffill['B'] = df_ffill['B'].ffill()

print("\033[96m" + "=== Forward Fill Only Column 'B' ===" + "\033[0m")
print(df_ffill, "\n")

# ============================================================
# Define a Smart Fill Function
#   - Numeric Columns  Fill with mean
#   - String Columns   Fill with 'Unknown'
# ============================================================
def fill_missing(df):
    for col in df.columns:
        if df[col].dtype == 'O':  # Object/String column
            df[col].fillna('Unknown', inplace=True)
        else:  # Numeric column
            df[col].fillna(df[col].mean(), inplace=True)
    return df

df_filled_func = fill_missing(df.copy())

print("\033[92m" + "=== Smart Fill Function Result ===" + "\033[0m")
print(df_filled_func, "\n")
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})

print("\033[95m" + "=== Original DataFrame ===" + "\033[0m")
print(df, "\n")

# ============================================================
# Drop rows where column 'A' has missing values
# ============================================================
df_drop_A = df.dropna(subset=['A'])
print("\033[94m" + "=== Drop Rows Where Column 'A' Has Missing Values ===" + "\033[0m")
print(df_drop_A, "\n")

# ============================================================
# Drop rows with less than 2 non-null values
# ============================================================
df_thresh = df.dropna(thresh=2)
print("\033[96m" + "=== Drop Rows with Less Than 2 Non-Null Values ===" + "\033[0m")
print(df_thresh, "\n")

# ============================================================
# Drop rows where BOTH Columns 'A' and 'B' are missing
# ============================================================
df_drop_AB_all = df.dropna(subset=['A', 'B'], how='all')
print("\033[92m" + "=== Drop Rows Where Both 'A' and 'B' Are Missing ===" + "\033[0m")
print(df_drop_AB_all, "\n")
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Value1': [10, 20, 30, 40, 50, 60],
    'Value2': [100, 200, 300, 400, 500, 600]
})

print("\033[95m" + "=== Original DataFrame ===" + "\033[0m")
print(df, "\n")

# ============================================================
# Mean of Value1 for each Category
# ============================================================
print("\033[94m" + "Mean of Value1 for each Category:" + "\033[0m")
print(df.groupby('Category')['Value1'].mean(), "\n")

# ============================================================
# Max of Value2 for each Category
# ============================================================
print("\033[96m" + "Max of Value2 for each Category:" + "\033[0m")
print(df.groupby('Category')['Value2'].max(), "\n")

# ============================================================
# Sum of Value1 for each Category
# ============================================================
print("\033[92m" + "Sum of Value1 for each Category:" + "\033[0m")
print(df.groupby('Category')['Value1'].sum(), "\n")

# ============================================================
# Multiple Aggregations using .agg()
# ============================================================
print("\033[93m" + "Multiple Aggregations (mean, sum, max):" + "\033[0m")
result = df.groupby('Category').agg({
    'Value1': ['mean', 'sum', 'max'],
    'Value2': ['mean', 'sum', 'min']
})
print(result, "\n")
```

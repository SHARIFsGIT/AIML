```python
string = "This is a new exercise for testing frequency of different words Lets see how can we count this If we can count it perfectly then it will be the best"

same_str = string.lower()

str_split = same_str.split()

count = {}

for word in str_split:
  count[word] = count.get(word, 0) + 1

print(count)
```

[J. Count Letters](https://codeforces.com/group/MWSDmqGsZm/contest/219856/problem/J)

```python
string = input()

count = { }

for char in string:
  count[char] = count.get(char, 0) + 1

for key in sorted(count):
  print(f"{key} : {count[key]}")
```

[No Idea!](https://www.hackerrank.com/challenges/no-idea/problem)

```python
nm = input().split()

n = int(nm[0])
m = int(nm[1])

arr = input().split()
int_arr = []

for i in arr:
  int_arr.append(int(i))

a = input().split()
a_set = set()

for i in a:
  a_set.add(int(i))

b = input().split()
b_set = set()

for i in b:
  b_set.add(int(i))

happiness = 0

for i in int_arr:
  if i in a_set:
    happiness += 1
  elif i in b_set:
    happiness -= 1

print(happiness)
```
